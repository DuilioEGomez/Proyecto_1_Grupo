```python
ruta_de_archivos = '/content/drive/MyDrive/Colab Notebooks/Laboratorio 1'
archivos = os.listdir(ruta_de_archivos)

dataframes= {}

for archivo in archivos:
    if archivo.endswith('.csv'):  # Asegurarse de que solo se consideren archivos CSV
        mes = archivo.split('_')[3].split('.')[0]  # Obtener el nombre del archivo sin la extensión
        ruta_completa = os.path.join(ruta_de_archivos, archivo)  # Ruta completa al archivo
        df = pd.read_csv(ruta_completa)  # Cargar el archivo CSV en un DataFrame
        df_ventas = df.copy()

        df_ventas['Cantidad Pedida'] = pd.to_numeric(df_ventas['Cantidad Pedida'], errors='coerce')  #El errors= 'coerce' transforma los valores no numericos a NaN'
        df_ventas['Precio Unitario'] = pd.to_numeric(df_ventas['Precio Unitario'], errors='coerce')


        df_ventas = df_ventas.loc[~df_ventas['Cantidad Pedida'].isna()]
        df_ventas = df_ventas.loc[~df_ventas['Precio Unitario'].isna()]

        # Elimino todos los Nan y me quedo solo con los valores numericos
        df_ventas = df_ventas.dropna()

        # Transformo todos los valores numericos al tipo de dato int
        df_ventas['Cantidad Pedida'] = df_ventas['Cantidad Pedida'].astype(int)
        df_ventas['Precio Unitario'] = df_ventas['Precio Unitario'].astype(int)

        dataframes[mes] = df_ventas  # Agregar el DataFrame al diccionario usando el nombre del mes
```

**5. Tendencia ventas:**

● ¿Existe alguna tendencia o patrón en las ventas que se repita a lo largo de los días en los diferentes meses?

● ¿Cómo varían las ventas a lo largo de los diferentes días de la semana?

● ¿Se observa alguna diferencia significativa en las ventas entre los días laborables y los fines de semana?

```python
from datetime import datetime
import calendar
def extraer_datos_Igna(datos):
    datos['Fecha de Pedido'] = pd.to_datetime(datos['Fecha de Pedido'])
    datos['Hora'] = datos['Fecha de Pedido'].dt.hour
    datos['Dia'] = datos['Fecha de Pedido'].dt.day
    datos['Dia de la Semana'] = datos['Fecha de Pedido'].dt.day_name()
    return datos

# Lista de nombres de los meses y los dataframes correspondientes
meses = ['Enero', 'Febrero', 'Marzo', 'Abril', 'Mayo', 'Junio', 'Julio', 'Agosto', 'Septiembre', 'Octubre', 'Noviembre', 'Diciembre']

# Diccionario para traducir los nombres de los días de la semana
traduccion_dias = {
    0: 'Lunes',
    1: 'Martes',
    2: 'Miércoles',
    3: 'Jueves',
    4: 'Viernes',
    5: 'Sábado',
    6: 'Domingo'
}

# Fecha de inicio conocida (primer día de enero de 2019)
fecha_conocida = datetime(2019, 1, 1)

# Dentro del bucle para cada mes
for mes in meses:
    dataframe_mes = dataframes[mes]
    dataframe_mes = extraer_datos_Igna(dataframe_mes)

    # Convierte la columna 'Fecha de Pedido' en formato de fecha
    dataframe_mes['Fecha de Pedido'] = pd.to_datetime(dataframe_mes['Fecha de Pedido'], format='%m/%d/%y %H:%M')

    # Calcula el total de unidades vendidas por día y el día de la semana
    dataframe_mes['Dia del Mes'] = dataframe_mes['Fecha de Pedido'].dt.day

    # Calcula el número máximo de días en el mes actual usando el módulo calendar
    mes_actual = meses.index(mes) + 1
    max_dia_mes = calendar.monthrange(2019, mes_actual)[1]

    # Agrupa por día del mes y suma las unidades vendidas
    ventas_por_dia_mes = dataframe_mes.groupby('Dia del Mes')['Cantidad Pedida'].sum().reset_index()

    # Filtra solo los días del mes actual
    ventas_por_dia_mes = ventas_por_dia_mes[ventas_por_dia_mes['Dia del Mes'] <= max_dia_mes]

    # Crea el gráfico de barras con la escala de color 'RdYlGn'
    fig_mes = px.bar(ventas_por_dia_mes, x='Dia del Mes', y='Cantidad Pedida',
                     title=f'Ventas por día en {mes}', labels={'Dia del Mes': 'Día del Mes', 'Cantidad Pedida': 'Unidades Vendidas'},
                     color='Cantidad Pedida', color_continuous_scale='RdYlGn')

    # Calcula los nombres de los días de la semana dinámicamente
    dias_semana = []

    # Obtén el día de la semana para el primer día del mes actual
    primer_dia_mes = datetime(2019, mes_actual, 1).weekday()

    # Llena la lista de nombres de días para el mes actual
    for dia in range(1, max_dia_mes + 1):
        dia_semana = (primer_dia_mes + dia - 1) % 7
        nombre_dia = traduccion_dias[dia_semana]
        dias_semana.append(nombre_dia)

    # Configura el texto dentro de las barras con el nombre del día y el total de ventas
    text_labels = [f'{nombre_dia}<br>{cantidad}' for nombre_dia, cantidad in zip(dias_semana, ventas_por_dia_mes['Cantidad Pedida'])]
    fig_mes.update_traces(text=text_labels, textposition='inside', textangle=90, showlegend=False)

    # Establece valores específicos en el eje x y muestra el gráfico
    fig_mes.update_xaxes(tickvals=list(range(1, max_dia_mes + 1)), ticktext=list(range(1, max_dia_mes + 1)))
    fig_mes.show()
```
